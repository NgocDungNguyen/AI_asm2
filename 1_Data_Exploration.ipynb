```python
# 1_Data_Exploration.ipynb

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'data/weather.xlsx'
weather_data = pd.read_excel(file_path)

# Display the first few rows of the dataset
print(weather_data.head())

# Summary statistics
print(weather_data.describe())

# Check for missing values
print(weather_data.isnull().sum())

# Histogram of max temperature
plt.figure(figsize=(10, 6))
sns.histplot(weather_data['max'], kde=True)
plt.title('Distribution of Max Temperature')
plt.xlabel('Max Temperature')
plt.ylabel('Frequency')
plt.show()

# Pairplot for numerical features
sns.pairplot(weather_data)
plt.show()

# Correlation matrix (for numerical features only)
plt.figure(figsize=(12, 8))
numeric_data = weather_data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Boxplot for detecting outliers in numerical features
plt.figure(figsize=(12, 6))
weather_data.boxplot(column=['max', 'min', 'wind', 'rain', 'humidi', 'cloud', 'pressure'])
plt.title('Boxplot for Numerical Features')
plt.show()
```
